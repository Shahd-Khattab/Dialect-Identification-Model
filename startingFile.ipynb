```python
# Importing the important libraries
import pandas as pd 
import numpy as np
```

```python
# Reading the data
data= pd.read_csv("transformed.csv")
data
```

```python
# Data type of each column
data.dtypes
```

```python
# Missing values in the data
nullData= data.isnull().sum()
nullData
```

```python
# Dropping the null values in the data
data= data.dropna(axis=0)
# Ensuring that there are no missing values
data.isnull().sum() / data.shape[0] * 100
```

```python
# Taking data sample of 1000 record from each dialect 
unique_dialects = data['dialect'].unique()

sampled_data = pd.DataFrame()

for dialect in unique_dialects:
    dialect_data = data[data['dialect'] == dialect]
    
    sampled_dialect_data = dialect_data.sample(n=1000, random_state=42)  
    
    sampled_data = pd.concat([sampled_data, sampled_dialect_data])
sampled_data
```

```python
# Saving the cleaned data as csv
sampled_data.to_csv('sampleData.csv')
```
